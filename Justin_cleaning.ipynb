# AA-AZ cleaning

### AC, AD, AE, AF, AG, AR, AH, AV, AT, AZ, BA-BB Dropping

```python
### Dropping BA-BB
Dropped_list = ['If the tenant was given prior notice for the eviction, how much notice was given?',
                'Exec Review', 
                'Review Status',
                'If yes to the previous question, did the decision state these conditions would make moving particularly burdensome?', # AH
                'Did the decision state any of the children had mental, medical or physical conditions?', # AG
                'How many total children aged 4 or younger did the tenant have living with them?', # AF
                'How many total children aged 13 or younger did the tenant have living with them? ', # AE
                'How many total children aged 17 or younger did the tenant have living with them?', # AD
                'How many total children did the tenant have living with them? ', # AC
                'If a payment plan was ordered, what was the length of the payment plan? ', # AR
                'If yes to the previous question, which of the following were applicable to the tenant?', # AT
                'Were there detail(s) in the decision not captured by this questionnaire that should be included?' # AZ
                ]
# data = data.loc[:,'What were the specific mental, medical, or physical conditions of the tenant, if any? ':'Were there detail(s) in the decision not captured by this questionnaire that should be included?']
data = data.drop(columns = Dropped_list)
data
```

### AX cleaning(notice that if there isn't any other application, it will be an empty list)

```python
import re
def find_application_code(string):
    '''
    based on a given string, return a list of application code extracted from it
    '''
    regex_ax = '[A-Za-z][0-9](?=:)'
    application_code = re.findall(regex_ax, string)
    return application_code
# example = 'L1: Application to Evict a Tenant for Non-payment of Rent and to Collect Rent the Tenant Owes, L2: Application to End a Tenancy and Evict a Tenant, N5: Notice to End your Tenancy for Interfering with Others, Damage or Overcrowding'
data['Which other specific applications of the landlord or the tenant were mentioned?'] = data['Which other specific applications of the landlord or the tenant were mentioned?'].apply(find_application_code)
```

### Binary Cleaning Function

```python
### AY cleaning function, treat nan as NO
def change_to_binary(string):
    if not isinstance(string,str):
        return 0
    elif string.lower() == 'yes':
        return 1
    else:
        return 0
```

### AB, AL, AN, AS, AU, AW cleaning

```python
# AB
data['Did the decision state that the tenant had children living with them?'] = data['Did the decision state that the tenant had children living with them?'].apply(change_to_binary)
# AL
data['Did the member find the tenant had sufficient income to pay rent?'] = data['Did the member find the tenant had sufficient income to pay rent?'].apply(change_to_binary)
# AN
data['Did the decision mention the tenant lost their job leading up to or during the period of the hearing?'] = data['Did the decision mention the tenant lost their job leading up to or during the period of the hearing?'].apply(change_to_binary)
# AS
data['Did the decision mention the tenant’s difficulty finding alternative housing for any reason e.g.physical limitations, reliance on social assistance, etc.?'] = data['Did the decision mention the tenant’s difficulty finding alternative housing for any reason e.g.physical limitations, reliance on social assistance, etc.?'].apply(change_to_binary)
# AU
data['Did the decision state the tenant was given prior notice for the eviction?'] = data['Did the decision state the tenant was given prior notice for the eviction?'].apply(change_to_binary)
# AW
data['Did the decisions state postponement would result in the tenant accruing additional arrears?'] = data['Did the decisions state postponement would result in the tenant accruing additional arrears?'].apply(change_to_binary)
# AY
data['Did the decision mention the validity of an N4 eviction notice?'] = data['Did the decision mention the validity of an N4 eviction notice?'].apply(change_to_binary)
```

### AO cleaning

```python
def change_to_binary_AO(string):
    # there are lots of weird things like "not stated ." & "not stated. ", the following code already took care of them
    if ('not stated' in string.lower().strip() or string.lower().strip() == 'no'):
        return 0
    else:
        return 1
```

```python
data['Did the decision mention any other extenuating circumstances experienced by the tenant leading up to or during the period of the claim (e.g. hospitalization, death in the family, etc.)?'] = data['Did the decision mention any other extenuating circumstances experienced by the tenant leading up to or during the period of the claim (e.g. hospitalization, death in the family, etc.)?'].apply(change_to_binary_AO)
```

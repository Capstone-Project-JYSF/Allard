```python
import re
def new_o(x):
    def avg_o(col):
        total = 0
        count = 0
        for value in col:
            if "tated" not in value and "N" not in value:
                    if value.isdigit() or "." in value:
                        total += float(value)
                        count += 1
                    else:
                        first = re.findall(r'\d+', value)[0]
                        total += float(first)
                        count += 1
        average = total / count
        return average
    
    mean_o = avg_o(data.iloc[:,15])
    
    if "tated" in x or "N" in x:
        return mean_o
    elif x.isdigit() or "." in x:
        return x
    else:
        return re.findall(r'\d+', x)[0]
```

```python
def new_p(x):
    def avg_p(col):
        total = 0
        count = 0
        for value in col:
            value = value.lstrip("$").replace(",", "")
            if "tated" not in value:
                if "N" not in value:
                    total += float(value)
                    count += 1
        average = total / count
        return average
    
    mean_p = avg_p(data.loc[:,col_p])
    
    if "tated" in x or "N" in x:
        return mean_p
    else:
        return x.lstrip("$").replace(",", "")
```
